```python
from Bio import pairwise2
```

```python
from Bio.pairwise2 import format_alignment
```

```python
global_Q1_X = "GTGTTGG"
global_Q1_Y = "CAGCCGT"
global_Q2_X = "GCTTTCC"
global_Q2_Y = "TTCCGGT"
```

```python
alignments__global_Q1 = pairwise2.align.globalms(global_Q1_X,global_Q1_Y,9,-5,-4,0)
alignments__global_Q2 = pairwise2.align.globalms(global_Q2_X,global_Q2_Y,10,-4,-6,0)
```

```python
print("Global Q1")
for a in alignments__global_Q1:
  print(format_alignment(*a))

print("Global Q2")
for a in alignments__global_Q2:
  print(format_alignment(*a))
```

```python
local_Q1_X = "ACCGGCTA"
local_Q1_Y = "AGGCTAATT"
local_Q2_X = "GCTTTCC"
local_Q2_Y = "TTCCGGT"
```

```python
alignments__local_Q1 = pairwise2.align.localms(local_Q1_X,local_Q1_Y,7,-4,-3,0)
alignments__local_Q2 = pairwise2.align.localms(local_Q2_X,local_Q2_Y,10,-4,-6,0)
```

```python
print("Local Q1")
for a in alignments__local_Q1:
  print(format_alignment(*a))
print("Local Q2")
for a in alignments__local_Q2:
  print(format_alignment(*a))
```
